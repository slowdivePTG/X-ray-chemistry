```python
j = s[:25].copy()
j.append('  write(66, \'(E15.5)\') ' + ' tt\n')
i = 32
idx = 0
while i < len(s):
    if '!' in s[i]:
        j.append('\n')
        j.append('\n' + s[i])
        i += 1
    elif 'dn(idx' in s[i]:
        idx += 1
        print(idx, s[i].replace(' = &', ''))
    elif 'k(' in s[i]:
        import re
        num = re.search(r"\d+\.?\d*", s[i])[0]
        j.append('  write(66, \'(i3, i4, E15.5)\') {}, {}, {}\n'.format(
            idx, num, s[i][:-1].replace(' &', '').replace(' ', '')))
    
    i += 1
```

```python
j.append(s[-1])
with open('Trace.f90', 'w') as f:
    f.writelines(j)
```
