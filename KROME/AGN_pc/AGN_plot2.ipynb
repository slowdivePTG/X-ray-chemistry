## Default Model

```python
def plot_default(spe,
                 div=None,
                 d=4,
                 stop=True,
                 ax=None,
                 color=None,
                 label=None,
                 Range=[1e0, 2e7],
                 Dt=True,
                 logT=True):
    if ax == None:
        f, ax = plt.subplots(figsize=(12, 8))
    if label == None:
        label = r'${}$'.format(latex_name(spe))

    j = str(d)
    Abu1 = Abu(j)
    if div != None:
        Abu1.Div(div)
    Abu1.plot(ax=ax, spe=spe, label='X-ray', color=color, Dt=Dt, linewidth=3)
    ax.set_ylabel(r'$n(${}$)$'.format(label) + '/' + Abu1.name_div,
                  fontsize=40)
    if stop:
        j = '0' + str(d)
        Abu0 = Abu(j)
        if div != None:
            Abu.Div(div)
        Abu0.plot(ax=ax,
                  spe=spe,
                  color=color,
                  linestyle='--',
                  label='X-ray removed',
                  Dt=Dt,
                  linewidth=3)

    raw = np.loadtxt('./data/2dis_inf')

    d = r'$n_{\mathrm{H}}$'
    abu = raw[:, eval('krome_idx_' + spe)]
    if div != None:
        abu /= raw[:, eval('krome_idx_' + div)]
        d = 'n' + div
    if Dt:
        t = raw[:, 0] - 1e6
        ax.set_xlabel(r'$t-10^6$ (yr)', fontsize=40)
    else:
        t = raw[:, 0]
        ax.set_xlabel(r'$t$ (yr)', fontsize=40)
    ax.loglog(t,
              abu,
              label='No X-ray',
              color='black',
              linestyle='-.',
              linewidth=3)
    ax.legend(loc=0, prop={'size': 20})
    ax.set_xlim(Range)
```

## Galactic Distribution

```python
def plot_dis(spe, div=None, stop=True, ax=None, color=None, label=None):
    if ax == None:
        f, ax = plt.subplots(figsize=(12, 8))
    if label == None:
        label = r'${}$'.format(latex_name(spe))
    for i, d in enumerate([1, 2, 4, 8]):
        j = str(d)
        Abu1 = Abu(j)
        if div != None:
            Abu1.Div(div)

        Abu1.plot(ax=ax,
                  spe=spe,
                  label='{} kpc'.format(d),
                  color=color[i],
                  linewidth=3)
        ax.set_ylabel(r'$n(${}$)$'.format(label) + '/' + Abu1.name_div,
                      fontsize=40)
        ax.set_xlim([1e0, 2e7])
        if stop:
            j = '0' + str(d)
            Abu0 = Abu(j)
            if div != None:
                Abu.Div(div)
            Abu0.plot(ax=ax,
                      spe=spe,
                      color=color[i],
                      linestyle='--',
                      label=None,
                      linewidth=3)

    raw = np.loadtxt('./data/2dis_inf')
    t = raw[:, 0] - 1e6
    d = r'$n_{\mathrm{H}}$'
    abu = raw[:, eval('krome_idx_' + spe)]
    if div != None:
        abu /= raw[:, eval('krome_idx_' + div)]
        d = 'n' + div
    ax.loglog(t,
              abu,
              label='No X-ray',
              color='black',
              linestyle='-.',
              linewidth=3)
    ax.legend(loc=0, prop={'size': 20})
```

```python
f, ax = plt.subplots(3, 2, figsize=(24, 24), sharex=True)
ax = ax.flatten()
c_dis = ['#d7191c', '#fdae61', '#abd9e9', '#2c7bb6']

plot_dis('H2O', ax=ax[0], color=c_dis)
plot_dis('CH3OH', ax=ax[2], color=c_dis)
plot_dis('H2CO', ax=ax[4], color=c_dis)

plot_dis('H_DUSTH_DUSTO_DUST',
         ax=ax[1],
         color=c_dis,
         label=r'${}$ (dust)'.format(latex_name('H2O')))
plot_dis('C_DUSTH_DUSTH_DUSTH_DUSTO_DUSTH_DUST',
         ax=ax[3],
         color=c_dis,
         label=r'${}$ (dust)'.format(latex_name('CH3OH')))
plot_dis('H_DUSTH_DUSTC_DUSTO_DUST',
         ax=ax[5],
         color=c_dis,
         label=r'${}$ (dust)'.format(latex_name('H2CO')))
ax[4].set_xlabel(r'$t-10^6$ (yr)', fontsize=40)
ax[5].set_xlabel(r'$t-10^6$ (yr)', fontsize=40)
#ax[0].set_ylim([1e-10, 1e-9])
ax[1].set_ylim([1e-5, 3e-4])
ax[2].set_ylim([1e-19, 5e-10])
ax[3].set_ylim([1e-11, 2e-5])
#ax[4].set_ylim([1e-12, 1e-10])
ax[5].set_ylim([1e-11, 2e-5])
for a in ax:
    a.tick_params(labelsize=35)
plt.tight_layout()
plt.savefig('../../../Report/Fig/galactic_distribution_2.eps')
```

## Different depths inside a molecular cloud

```python
def plot_col(spe,
             div=None,
             d=4,
             stop=True,
             ax=None,
             color=['#377eb8', '#e41a1c', '#4daf4a'],
             label=None):
    if ax == None:
        f, ax = plt.subplots(figsize=(12, 8))
    if label == None:
        label = r'${}$'.format(latex_name(spe))

    j = str(d)
    Abu1 = Abu(j)
    ax.set_ylabel(r'$n(${}$)$'.format(label) + '/' + Abu1.name_div,
                  fontsize=40)
    ax.set_xlim([1e0, 2e7])
    if div != None:
        Abu1.Div(div)
    Abu1.plot_l(ax=ax,
                spe=spe,
                color=color[1],
                Nolabel=False,
                linestyle='-',
                linewidth=3)
    Abu1.plot_u(ax=ax,
                spe=spe,
                color=color[0],
                Nolabel=False,
                linestyle='-',
                linewidth=3)
    if stop:
        j = '0' + str(d)
        Abu0 = Abu(j)
        if div != None:
            Abu.Div(div)
        Abu0.plot_l(ax=ax,
                    spe=spe,
                    color=color[1],
                    Nolabel=False,
                    linestyle='--',
                    linewidth=3,
                    stop=True)
        Abu0.plot_u(ax=ax,
                    spe=spe,
                    color=color[0],
                    Nolabel=False,
                    linestyle='--',
                    linewidth=3,
                    stop=True)

    Abu_inf = Abu('_inf')
    Abu_inf.plot_l(ax=ax,
                   spe=spe,
                   color=color[1],
                   Nolabel=False,
                   NoX=True,
                   linestyle=':',
                   linewidth=5)
    Abu_inf.plot_u(ax=ax,
                   spe=spe,
                   color=color[0],
                   Nolabel=False,
                   NoX=True,
                   linestyle=':',
                   linewidth=5)
    ax.legend(loc=0, prop={'size': 20})
```

```python
f, ax = plt.subplots(3, 2, figsize=(24, 24), sharex=True)
ax = ax.flatten()

plot_col('H2O', ax=ax[0])
plot_col('CH3OH', ax=ax[2])
plot_col('H2CO', ax=ax[4])

plot_col('H_DUSTH_DUSTO_DUST',
         ax=ax[1],
         label=r'${}$ (dust)'.format(latex_name('H2O')))
plot_col('C_DUSTH_DUSTH_DUSTH_DUSTO_DUSTH_DUST',
         ax=ax[3],
         label=r'${}$ (dust)'.format(latex_name('CH3OH')))
plot_col('H_DUSTH_DUSTC_DUSTO_DUST',
         ax=ax[5],
         label=r'${}$ (dust)'.format(latex_name('H2CO')))
ax[4].set_xlabel(r'$t-10^6$ (yr)', fontsize=40)
ax[5].set_xlabel(r'$t-10^6$ (yr)', fontsize=40)
#ax[0].set_ylim([3e-12, 3e-5])
#ax[1].set_ylim([1e-7, 2e-4])
#ax[2].set_ylim([1e-23, 2e-5])
#ax[3].set_ylim([5e-11, 2e-5])
#ax[4].set_ylim([1e-14, 8e-6])
#ax[5].set_ylim([5e-10, 2e-5])
for a in ax:
    a.tick_params(labelsize=35)
plt.tight_layout()
plt.savefig('../../../Report/Fig/column_density_2.eps')
```

## Age of Molecular Clouds

```python
def plot_age(
        spe,
        div=None,
        Age={
            #'': 0,
            #'1e+04': '$10^4$ yr',
            '1e+07': '$10^7$ yr'
        },
        ax=None,
        color=None,
        label=None,
        Range=[1e0, 2e7],
        Dt=True):
    if ax == None:
        f, ax = plt.subplots(figsize=(12, 8))
    if label == None:
        label = r'${}$'.format(latex_name(spe))
    for i, age in enumerate(Age.keys()):
        j = '4{}'.format(age)
        Abu1 = Abu(j)
        if div != None:
            Abu1.Div(div)
        Abu1.plot(ax=ax,
                  spe=spe,
                  label='X-ray, Age = {}'.format(Age[age]),
                  color=color,
                  linewidth=3,
                  Dt=Dt)
        ax.set_ylabel(r'$n(${}$)$'.format(label) + '/' + Abu1.name_div,
                      fontsize=40)
        ax.set_xlim(Range)
        j = '04{}'.format(age)
        Abu0 = Abu(j)
        if div != None:
            Abu.Div(div)
        Abu0.plot(ax=ax,
                  spe=spe,
                  color=color,
                  linestyle='--',
                  label='X-ray removed, Age = {}'.format(Age[age]),
                  linewidth=3,
                  Dt=Dt)

        raw = np.loadtxt('./data/2dis_inf{}'.format(age))
        if Dt:
            t = raw[:, 0] - 1e6
        else:
            t = raw[:, 0]
        d = r'$n_{\mathrm{H}}$'
        abu = raw[:, eval('krome_idx_' + spe)]
        if div != None:
            abu /= raw[:, eval('krome_idx_' + div)]
            d = 'n' + div
        ax.loglog(t,
                  abu,
                  label='No X-ray, Age = {}'.format(Age[age]),
                  color='black',
                  linestyle='-.',
                  linewidth=3)
        ax.legend(loc=0, prop={'size': 20})
```

```python
f, ax = plt.subplots(3, 2, figsize=(24, 24), sharex=True)
ax = ax.flatten()
c_default = ['#7fc97f', '#beaed4', '#fdc086']

plot_age('H2O', ax=ax[0], color=c_default[0])
plot_age('CH3OH', ax=ax[2], color=c_default[1])
plot_age('H2CO', ax=ax[4], color=c_default[2])

plot_age('H_DUSTH_DUSTO_DUST',
         ax=ax[1],
         color=c_default[0],
         label=r'${}$ (dust)'.format(latex_name('H2O')))
plot_age('C_DUSTH_DUSTH_DUSTH_DUSTO_DUSTH_DUST',
         ax=ax[3],
         color=c_default[1],
         label=r'${}$ (dust)'.format(latex_name('CH3OH')))
plot_age('H_DUSTH_DUSTC_DUSTO_DUST',
         ax=ax[5],
         color=c_default[2],
         label=r'${}$ (dust)'.format(latex_name('H2CO')))
ax[4].set_xlabel(r'$t-10^6$ (yr)', fontsize=40)
ax[5].set_xlabel(r'$t-10^6$ (yr)', fontsize=40)
ax[1].set_ylim([3.54e-5, 3.6e-5])
#ax[2].set_ylim([1e-18, 1e-10])
ax[3].set_ylim([1.9e-8, 2.6e-8])
ax[5].set_ylim([5e-8, 8e-8])
    
for i, a in enumerate(ax):
    if i%2==1:
        a.set_yscale('linear')
        a.yaxis.get_major_formatter().set_powerlimits((0,1))
        tx = a.yaxis.get_offset_text() 
        tx.set_fontsize(25) 
    a.tick_params(labelsize=35, which='both')
plt.tight_layout()
plt.savefig('../../../Report/Fig/age_2.eps')
plt.show()
```
