```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from species import *
```

```python
def plot_species(txt,name):
    print(name)
    i = 0
    data = np.loadtxt(txt)
    log_jx21 = data[:,0]
    log_name = data[:,eval("krome_idx_"+name)]
    rg = linear_model.LinearRegression()
    rg.fit(log_jx21.reshape(-1,1),log_name)
    print(r'alpha=',rg.coef_[0])
    
    plt.scatter(10**log_jx21,log_name)
    plt.xlim([10**log_jx21[0]*0.995,10**log_jx21[-1]/0.995])
    plt.ylim([log_name.min()*1.0002,log_name.max()/1.0002])
    plt.xlabel(r"$Jx21$")
    plt.ylabel(r"$\log\ x/n_H$")
    plt.show()
```

```python
plot_species("case3","NO")
plot_species("case3","HCN")
plot_species("case3","HNOj")
plot_species("case3","HCNHj")
plot_species("case3","C2H2j")
plot_species("case3","HCNj")
```
