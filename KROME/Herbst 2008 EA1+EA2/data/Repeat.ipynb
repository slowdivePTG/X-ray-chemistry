# Repeat the results in Wakelam & Herbst 2008

```python
import numpy as np
import matplotlib.pyplot as plt
from species import *
```

```python
def plot_fig(txt, spe, div = None):
    raw = np.loadtxt(txt)
    t = raw[:,0]
    d = 'xH'
    abu = raw[:,eval('krome_idx_' + spe)]
    if div != None:
        abu /= raw[:,eval('krome_idx_' + div)]
        d = 'n' + div
    plt.loglog(t, abu, label = txt)
    plt.xlabel(r'$t$ (yr)')
    plt.ylabel(r'$n$'+spe.replace('j','+').replace('k','-')+'/'+d)
    plt.legend()
```

```python
species_list = ['C','Cj','H3j','HCOj','H3Oj','OH','H2O','O2']
for i in species_list:
    plot_fig('case1', i)
    plot_fig('case2', i)
    plt.show()
```

```python
species_list = ['C3H2','HC3N']
for i in species_list:
    plot_fig('case1', i)
    plot_fig('case2', i)
    plt.show()
```
