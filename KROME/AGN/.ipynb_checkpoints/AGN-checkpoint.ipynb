```python
Col = ['#b2182b','#d6604d','#f4a582','#fddbc7','#d1e5f0','#92c5de','#4393c3','#2166ac']
def plot_fig(txt, col, spe, div = None):
    raw = np.loadtxt(txt)
    t = raw[:,0]
    d = 'xH'
    abu = raw[:,eval('krome_idx_' + spe)]
    if div != None:
        abu /= raw[:,eval('krome_idx_' + div)]
        d = 'n' + div
    plt.loglog(t, abu, label = txt[-1]+' kpc', color = col)
    plt.xlabel(r'$t$ (yr)')
    plt.ylabel(r'$n$'+spe.replace('j','+').replace('k','-')+'/'+d)
    plt.legend()
```

```python
for i in range(8):
    dis = i+1
    txt = './data/dis'+str(dis)
    plot_fig(txt, Col[i], 'E')
plt.show()
```
