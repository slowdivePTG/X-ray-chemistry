# Repeat the results in Wakelam & Herbst 2008

```python
import numpy as np
import matplotlib.pyplot as plt
from species import *
```

```python
def plot_fig(txt, spe, div = None):
    raw = np.loadtxt(txt)
    t = raw[:,0]
    d = 'xH'
    abu = raw[:,eval('krome_idx_' + spe)]
    if div != None:
        abu /= raw[:,eval('krome_idx_' + div)]
        d = 'n' + div
    plt.loglog(t, abu, label = txt)
    plt.ylabel(r'$n($'+spe.replace('j','+').replace('k','-')+')/'+d)
    plt.legend()
```

```python
plt.figure(figsize=(8,3))
species_list = ['C','Cj']
j = 1
for i in species_list:
    plt.subplot(1,2,j)
    j += 1
    plot_fig('EA1', i)
    plot_fig('EA2', i)
    plot_fig('EA3', i)
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
plt.legend()
plt.savefig('1.eps')
```

```python
plt.figure(figsize=(8,9))
species_list = ['H3j','HCOj','H3Oj','OH','H2O','O2']
plt.subplot(3,2,1)
plot_fig('EA1', 'H3j')
plot_fig('EA2', 'H3j')
plot_fig('EA3', 'H3j')
plt.xlim([1e2,1e7])
plt.ylim([1e-11,10**(-8.5)])

plt.subplot(3,2,2)
plot_fig('EA1', 'HCOj')
plot_fig('EA2', 'HCOj')
plot_fig('EA3', 'HCOj')
plt.xlim([1e2,1e7])
plt.ylim([1e-12,1e-8])

plt.subplot(3,2,3)
plot_fig('EA1', 'H3Oj')
plot_fig('EA2', 'H3Oj')
plot_fig('EA3', 'H3Oj')
plt.xlim([1e2,1e7])
plt.ylim([1e-12,1e-8])

plt.subplot(3,2,4)
plot_fig('EA1', 'OH')
plot_fig('EA2', 'OH')
plot_fig('EA3', 'OH')
plt.xlim([1e2,1e7])
plt.ylim([1e-12,1e-7])

plt.subplot(3,2,5)
plot_fig('EA1', 'H2O')
plot_fig('EA2', 'H2O')
plot_fig('EA3', 'H2O')
plt.xlabel(r'$t$ (yr)')
plt.xlim([1e2,1e7])
plt.ylim([1e-11,1e-4])

plt.subplot(3,2,6)
plot_fig('EA1', 'O2')
plot_fig('EA2', 'O2')
plot_fig('EA3', 'O2')
plt.xlabel(r'$t$ (yr)')
plt.xlim([1e2,1e7])
plt.ylim([1e-11,1e-4])

plt.tight_layout(pad=0.4)
plt.savefig('2.eps')
```

```python
species_list = ['C3H2','HC3N']
for i in species_list:
    plot_fig('EA1', i)
    plot_fig('EA2', i)
    plot_fig('EA3', i)
    plt.show()
```
