# Repeat the results in Wakelam & Herbst 2008

```python
import numpy as np
import matplotlib.pyplot as plt
from species import *
```

```python
def plot_fig(txt, spe, div = None):
    raw = np.loadtxt(txt)
    t = raw[:,0]
    d = 'xH'
    abu = raw[:,eval('krome_idx_' + spe)]
    if div != None:
        abu /= raw[:,eval('krome_idx_' + div)]
        d = 'n' + div
    plt.loglog(t, abu, label = txt)
    plt.ylabel(r'$n($'+spe.replace('j','+').replace('k','-')+')/'+d)
    plt.legend()
```

```python
plt.figure(figsize=(8,3))
species_list = ['C','Cj']
j = 1
for i in species_list:
    plt.subplot(1,2,j)
    j += 1
    plot_fig('EA1', i)
    plot_fig('EA2', i)
    plot_fig('EA3', i)
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
plt.legend()
plt.savefig('1.eps')
```

```python
plt.figure(figsize=(8,9))
species_list = ['H3j','HCOj','H3Oj','OH','H2O','O2']
j = 1
for i in species_list:
    plt.subplot(3,2,j)
    j += 1
    plot_fig('EA1', i)
    plot_fig('EA2', i)
    plot_fig('EA3', i)
    if j>=len(species_list):
        plt.xlabel(r'$t$ (yr)')
plt.tight_layout(pad=0.4)
plt.savefig('2.eps')
```

```python
species_list = ['C3H2','HC3N']
for i in species_list:
    plot_fig('EA1', i)
    plot_fig('EA2', i)
    plot_fig('EA3', i)
    plt.show()
```
